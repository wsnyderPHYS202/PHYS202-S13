##Publication Ready Plot Templates
###Simple graphing template for matplotlib functions and data plots.

####(b) Line Plots

```python
#Building an array from where -5 < y < 5
y = arange(-4.99, 5, .01)

#Setting E and p to their values
E = cosh(y)
p = sinh(y)
```

```python
figure(figsize =(8,6)) #Resizes the plot

plot(E,p, color = "blue", label = "Name of line in Legend")#Plots E vs p with a color and label
#We can also alter many other aspects of the plot like line thickness and line style

#Setting the axis
axis([-100,100,-100,100]) #Formatted as axis([xmin,xmax,ymin,ymax])

#Centers the spines 
ax = gca()
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['right'].set_position('center')
ax.spines['top'].set_position('center')

#Moving the ticks to the opposite side
ax.xaxis.tick_top()
ax.yaxis.tick_right()

#Labeling axis, setting ticks
xlabel("$Label$ $for$ $X-axis$", fontsize = 20)
ylabel("$Label$ $for$ $Y-axis$", fontsize = 20)
xticks()
yticks()


#Makes grid, legend, and title
grid(True)
legend(loc = "upper left")
title("$Title$", fontsize = 20)
```

####(e) Subplots

```python
#Creates figure object
fig = figure()

#Makes the two subplots
first_sub = fig.add_subplot(2,1,1)
second_sub = fig.add_subplot(2,1,2)

#Plots the two subplots
first_sub.plot(E,p, color = "b", label = "Name of line in Legend")
second_sub.plot(E,p, color = "b", label = "Name of line in Legend")
```

```python
n = 200
x = random.normal(0,1,n)
y = random.normal(0,1,n)
```

```python
figure(figsize =(8,6)) #Resizes the plot

scatter(x,y, color = "burlywood", label = "Name of line in Legend")#Plots E vs p with a color and label
#We can also alter many other aspects of the plot like line thickness and line style

#Setting the axis
axis([-3,3,-3,3]) #Formatted as axis([xmin,xmax,ymin,ymax])

#Centers the spines 
ax = gca()
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['right'].set_position('center')
ax.spines['top'].set_position('center')

#Moving the ticks to the opposite side
ax.xaxis.tick_top()
ax.yaxis.tick_right()

#Labeling axis, setting ticks
xlabel("$Random$ $X$ $Values$", fontsize = 20)
ylabel("$Random$ $Y$ $Values$", fontsize = 20)
xticks()
yticks()

#Makes grid, legend, and title
grid(True)
legend(loc = "best")
title("$Random Plot$", fontsize = 20)
```
