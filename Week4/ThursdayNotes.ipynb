## Interpolation

```python
theta = arange(0,2.*pi + 0.1,pi/10)
mask = (theta == 0.5)
#print mask
```

```python
from scipy.interpolate import interp1d
#help(interp1d)
newTheta = 0.5; newSin = 0.
sinInterp = interp1d(theta, sin(theta))
newSin = sinInterp(newTheta)
print newSin
```

```python
plot(theta,sin(theta),newTheta,newSin, 'r+',markersize = 20, markeredgewidth = 3)
```

```python
thetaNew = arange(0,2*pi + 0.1, pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta), thetaNew, sinNew, 'r*', markersize = 10)
show()
```

```python
plot(thetaNew, sinNew - sin(thetaNew))
ylabel('difference betwen interp and exact value')
show()
```

```python
sinNewC = interp1d(theta, sin(theta), kind = 'cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew), thetaNew, sinNewC-sin(thetaNew), 'burlywood')
subplot(2,1,2)
plot(thetaNew, sinNewC-sin(thetaNew), 'chartreuse')
show()
```

```python
x = arange(0.,10.1,0.1)
y = x**5
```

```python
dy = diff(y)
#help(diff)
print len(y),len(dy)
dx = diff(x)
dydx = (dy/dx)
print len(dydx)
```

```python
plot(x,dydx)
```

```python
dydx1 = zeros(y.shape, float)
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
dydx2 = zeros(y.shape, float)
dydx2[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
dydx2[0] = (y[1]-y[0])/(x[1]-x[0])
dydx2[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
dydxExact = 5*x**4
```

```python
plot(x,dydx1,label = 'foreward difference')
plot(x,dydx2,label = 'center difference')
plot(x,dydxExact,label = 'exact')
legend(loc = 'upper left')
show()
```

```python
percentError1 = 100*abs(dydx1[1:]-dydxExact[1:])/dydxExact[1:]
percentError2 = 100*abs(dydx2[1:]-dydxExact[1:])/dydxExact[1:]

semilogy(x[1:], percentError1, label = 'foreward difference')
semilogy(x[1:], percentError2, label = 'center difference')
ylabel = 'Percent Error with dydxExact'
legend(loc = 'upper right')
show()
```
