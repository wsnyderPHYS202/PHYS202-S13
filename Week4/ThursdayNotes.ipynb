## Interpolation

```python
theta = arange(0,2.*pi + 0.1,pi/10)
mask = (theta == 0.5)
#print mask
```

```python
from scipy.interpolate import interp1d
#help(interp1d)
newTheta = 0.5; newSin = 0.
sinInterp = interp1d(theta, sin(theta))
newSin = sinInterp(newTheta)
print newSin
```

```python
plot(theta,sin(theta),newTheta,newSin, 'r+',markersize = 20, markeredgewidth = 3)
```

```python
thetaNew = arange(0,2*pi + 0.1, pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta), thetaNew, sinNew, 'r*', markersize = 10)
show()
```

```python
plot(thetaNew, sinNew - sin(thetaNew))
ylabel('difference betwen interp and exact value')
show()
```

```python
sinNewC = interp1d(theta, sin(theta), kind = 'cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew), thetaNew, sinNewC-sin(thetaNew), 'burlywood')
subplot(2,1,2)
plot(thetaNew, sinNewC-sin(thetaNew), 'chartreuse')
show()
```
