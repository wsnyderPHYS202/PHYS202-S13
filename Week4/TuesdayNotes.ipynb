# Electric Dipole Potential

$V(x,y) = \cfrac{kq}{\sqrt{x^2 + (y - d)^2}} - \cfrac{kq}{\sqrt{x^2 + (y + d)^2}}$

```python
dx = 0.06
dy = 0.06

x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))

k = 1.; d = 1.; q = 1.
V = (k*q)/sqrt((x**2) + (y-d)**2) - (k*q)/sqrt((x**2) + (y+d)**2)
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x, y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.coolwarm)

cb = fig.colorbar(p, shrink = 0.5)

cset = ax.contour(x, y, V, zdir = 'z', ofset = -40., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'x', ofset = -2., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'y', ofset = -2., cmap = cm.coolwarm)
```

```python
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])

contour(x,y,V,levels, linewidths = 2)
show()
```

## Ex. 3

#### (a)

```python
def pointPotential(x,y,q,posx,posy):
    """This function will take in the paramaters of position and charge and return the voltage due to that charge"""
    Vxy = (k*q)/sqrt((x-posx)**2 + (y-posy)**2)
    return Vxy
```

#### (b)

```python
def dipolePotential(x,y,q,d):
    """Returns the potential at (x,y) of a dipole with d being the distance between the charges and q being the charges"""
    V1 = pointPotential(x,y,q,-d/2,0.)
    V2 = pointPotential(x,y,q,d/2,0.)
    Vdp = V1 - V2
    return Vdp
```

#### (c)

```python
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
k = 1.

Vdp = dipolePotential(x,y,1.,1.)
```

```python
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x, y, Vdp, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.gist_earth_r)

cb = fig.colorbar(p, shrink = 0.5)

cset = ax.contour(x, y, Vdp, zdir = 'z', ofset = -40., cmap = cm.gist_earth_r)
cset = ax.contour(x, y, Vdp, zdir = 'x', ofset = -2., cmap = cm.gist_earth_r)
cset = ax.contour(x, y, Vdp, zdir = 'y', ofset = -2., cmap = cm.gist_earth_r)
```
