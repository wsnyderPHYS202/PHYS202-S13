# Electric Dipole Potential

$V(x,y) = \cfrac{kq}{\sqrt{x^2 + (y - d)^2}} - \cfrac{kq}{\sqrt{x^2 + (y + d)^2}}$

```python
dx = 0.06
dy = 0.06

x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))

k = 1.; d = 1.; q = 1.
V = (k*q)/sqrt((x**2) + (y-d)**2) - (k*q)/sqrt((x**2) + (y+d)**2)
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(x, y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.coolwarm)

cb = fig.colorbar(p, shrink = 0.5)

cset = ax.contour(x, y, V, zdir = 'z', ofset = -40., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'x', ofset = -2., cmap = cm.coolwarm)
cset = ax.contour(x, y, V, zdir = 'y', ofset = -2., cmap = cm.coolwarm)
```

```python
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])

contour(x,y,V,levels, linewidths = 2)
show()
```
