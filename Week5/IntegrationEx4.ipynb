```python
#Finding the number of slices
A = arange(0.,3.1,0.1)
A.size
print A
```

```python
#Calculating the integral using Simpson's Rule
f =lambda x: e**(-(x**2))
a = 0.0
slices = 30 
data = [0.0]
for i in A[1:]:
    h = abs(a-i)/slices
    loopNum = slices/2
    sum1 = 0
    sum2 = 0
        
    for k in range(1,loopNum + 1):
        sum1 += f(a+(2*k-1)*h)
        if (k < (loopNum)):
            sum2 += f(a +2*k*h)
    
    area = (h/3)*(f(a) + f(i) + 4*sum1 + 2*sum2)
    data = append(data,area)
print data
```

```python
#Calculating the integral using Trapezoidal Rule
f = lambda x: e**(-(x**2))
a = 0.0
N = A.size
for q in A:
    h = (q-a)/N
    s = 0.5*f(a)+0.5*f(q)
    for k in range(1,N):
        s += f(a+k*h)
    print (h*s)
```

```python
from math import *
```

```python
def H(n,x):
    if n == 0:
        herm = 1
    elif n == 1:
        herm = 2*x
    elif n == 2:
        herm = 4*(x**2) - 2
    else:
        zeroth = 1
        prev = 2*x
        current = 2*x*prev - 2*2*zeroth
        for q in range(2,n):
            herm = 2*x*current - 2*q*prev
            prev = current
            current = herm
    return herm
```

```python
psi_n = lambda x,n: ((e**(-(x**2)/2.))/sqrt((2**n)*factorial(n)*sqrt(pi)))*H(n,x)
```

```python
H(2,2)
```

###Problem 4

####(a)

```python
def MonteCarloMethod(N,xmin,xmax):
    f = lambda x: 4*np.sqrt(1-x**2)

    x = arange(xmin,xmax,0.001)
    
    subx = x[logical_and((x>=xmin),(x<=xmax))]
    ymax = max(f(subx))*1.05
    ymin = 0.

    samples_x = xmin + (xmax - xmin)*random_sample(N);
    samples_y = ymin + (ymax - ymin)*random_sample(N);

    newmask = (samples_y < f(samples_x))

    I = (xmax - xmin)*(ymax-ymin)* sum(newmask)/newmask.size
    return I
```

```python
result = MonteCarloMethod(1000,0.,1.) #Using N = 1,000
print result
```

####(b)

```python
err = 100*abs(pi - result)/pi
print 'Error:',err,'%'
```

####(c)

```python
N_arr = arange(1,100000,10)

blob = []
for q in N_arr:
    blob = append(blob, MonteCarloMethod(q,0.,1.))
    
figure(figsize = (8,6))
loglog(N_arr, 100*abs(pi - blob)/pi, label = 'percent error', color = 'b')
xlabel("$N$", fontsize = 20)
ylabel("$Percent Error$", fontsize = 20)

grid(True)
legend(loc = 'lower left')

show()
```

The plot shows that as the N gets larger the Error begins to vary greatly and the N doesn't matter as much. For example around $10^3$ we were getting some of the same Percent Error as around $10^5$.

```python
def SumMean(N,xmin,xmax):
    f = lambda x: 4*np.sqrt(1 - x**2)

    xi = linspace(xmin,xmax,N)

    b_a = xmax - xmin
    A = 1./N
    C = b_a * A

    tot = 0.
    for q in xi:
        tot += f(q)
    sumMean = C*tot

    return sumMean
```
