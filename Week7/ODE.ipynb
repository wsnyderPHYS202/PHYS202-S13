##Ordinary Differential Equations

```python
g = 9.8 #m/s^2
l = 0.5 #m
th0 = 20.*pi/180 #20 degrees in radians

t = arange(0.,10. + 0.01, 0.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)
    
plot(t,theta_app(t))
show()
```

```python
def deriv(theta_current, t):
    g = 9.8 #m/s^2
    l = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
from scipy.integrate import odeint
#odeint?
```

```python
t = arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = array([th0, omega0])

theta = odeint(deriv,theta_init,t) #pass the function deriv to the solver
```

```python
theta.shape
```

```python
plot(t,theta_app(t),'b-',t,theta[:,0],'r--')
show()
```

```python
th0 = 50*pi/180

theta_init = array([th0, omega0])
theta = odeint(deriv,theta_init,t) #pass the function deriv to the solver

plot(t,theta_app(t),'b-',t,theta[:,0],'r--')
show()
```

###Ex. 1

```python
theta_app0 = lambda t,th0: th0*cos(sqrt(g/l)*t)

th01 = 0
th02 = pi/4
th03 = pi/2
th04 = 3*pi/4
th05 = pi

theta_init1 = array([th01, omega0])
theta1 = odeint(deriv,theta_init1,t)

theta_init2 = array([th02, omega0])
theta2 = odeint(deriv,theta_init2,t)

theta_init3 = array([th03, omega0])
theta3 = odeint(deriv,theta_init3,t)

theta_init4 = array([th04, omega0])
theta4 = odeint(deriv,theta_init4,t)

theta_init5 = array([th05, omega0])
theta5 = odeint(deriv,theta_init5,t)

figure(figsize = (8,6))

plot(t,theta_app0(t,th01),'y-',t,theta1[:,0],'y--')# theta0 =0
plot(t,theta_app0(t,th02),'k-',t,theta2[:,0],'k--')# theta0 = pi/4
plot(t,theta_app0(t,th03),color = 'purple')
plot(t,theta3[:,0],color = 'purple', linestyle = '--')# theta0 = pi/2
plot(t,theta_app0(t,th04),'g-',t,theta4[:,0],'g--')# theta0 = 3pi/4
plot(t,theta_app0(t,th05),'b-',t,theta5[:,0],'b--')# theta0 = pi
show()
```

The approximation eror at pi/4 is minimal for the first 5 seconds. At 3pi/4 the error is aparent from almost the beggining and the period of the approximation is visibly different from the line. The trajectories of the 0 and pi inital angles are the same.
