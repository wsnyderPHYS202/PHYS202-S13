###Example 1

```python
x = arange(0.,pi+0.01,0.01)
f = lambda x : sin(x)/2.
F = lambda x : (cos(0) - cos(x))/2.

figure(figsize =(8,6))

plot(x,f(x), label = 'True distribution', lw = 3)
plot(x,F(x), label = 'Cumulative function', lw = 3)

xlim(0.,3.5)
ylim(0.,1.1)

legend(loc = 0)
show()
```

```python
N = 10000

z = numpy.random.random_sample(N);

Finv = lambda z: arccos(cos(0)-2*z)

figure(figsize = (8,6))

xsamples = Finv(z)
    
hist(xsamples,50,normed = True,label = 'Samples distribution')
plot(x,f(x),label='True distribution',lw = 3, color = 'r')
         
xlim(0,pi)
ylim(0,0.7)

legend(loc = 0)
show()
```

###Exercise 2. Exponential Function - Numerical

####(a)

```python
x1 =arange(0.,10.1,0.1)
lam = 1
f = lambda x: lam*e**(-lam*x)
    
farr = f(x1)
#print farr

A = cumsum(farr)
#print A
B = sum(farr)
normA = A/B
#print normA

N = 100000
z = numpy.random.random_sample(N);
#print z

dex = []
for q in z:
    dex = append(dex,searchsorted(normA,q))
#print dex

thing = []
for i in dex:
    thing = append(thing,x1[i])
#print thing
```

####(b)

```python
figure(figsize = (8,6))
    
hist(thing,50,normed = True,label = 'Samples distribution')
plot(x1,f(x1),label='True distribution',lw = 3, color = 'r')

legend(loc = 0)
show()
```

####(c)

```python
def distripute1D(xmax,f,N):
    x =arange(0.001,xmax + 0.001,0.001)
    
    farr = f(x)
    #print farr
    A = cumsum(farr)
    B = sum(farr)
    normA = A/B

    z = numpy.random.random_sample(N);

    dex = []
    for q in z:
        dex = append(dex,searchsorted(normA,q))

    thing = []
    for i in dex:
        thing = append(thing,x[i])
    return thing
```

###Exercise 3 Quantum Double-Slit Experiment

```python
a = 10e-6
d = 15e-6
L = 1
lm = 632.8e-9
Ascreen = 0.2*0.2
```

```python
sx = pi*a/(lm*L)
cx = 2*pi*d/(lm*L)
IDStot = lambda x: sin(sx*x)**2/((sx*x)) * (cos(cx*x)**2)
    
photons = distripute1D(0.2,IDStot,10000)

print photons#, photons.size
```

####(b)

```python
xarray = arange(0.,0.2+0.001,0.001)

figure(figsize = (8,6))
    
n, bins, patches = hist(photons,50,normed=True,label="Sampled data")
plot(xarray,IDStot(xarray)*max(n),'r-',label='True distribution', lw = 2)

legend(loc = 0)
show()
```

The histogram fairly accurately describes the true distribution.
