###Example 1

```python
x = arange(0.,pi+0.01,0.01)
f = lambda x : sin(x)/2.
F = lambda x : (cos(0) - cos(x))/2.

figure(figsize =(8,6))

plot(x,f(x), label = 'True distribution', lw = 3)
plot(x,F(x), label = 'Cumulative function', lw = 3)

xlim(0.,3.5)
ylim(0.,1.1)

legend(loc = 0)
show()
```

```python
N = 10000

z = numpy.random.random_sample(N);

Finv = lambda z: arccos(cos(0)-2*z)

figure(figsize = (8,6))

xsamples = Finv(z)
    
hist(xsamples,50,normed = True,label = 'Samples distribution')
plot(x,f(x),label='True distribution',lw = 3, color = 'r')
         
xlim(0,pi)
ylim(0,0.7)

legend(loc = 0)
show()
```

###Exercise 1. Exponential Function - Analytic

####(a)

```python
f = lambda x,lam: lam*e**(-lam*x)
F = lambda x,lam: 1-e**(-lam*x)
Finv = lambda z,lam: -log(1-z)/lam
```

####(b)

```python
x = arange(0.,5+0.01,0.01)
lam = 1

figure(figsize =(8,6))

plot(x,f(x,lam), label = 'True distribution', lw = 3)
plot(x,F(x,lam), label = 'Cumulative function', lw = 3)

xlim(0.,5.)
ylim(0.,1.1)

legend(loc = 0)
show()
```

####(c)

```python
N = 10000
x = arange(0.,10+0.01,0.01)
z = numpy.random.random_sample(N);

Finv = lambda z,lam: -log(z)/lam

figure(figsize = (8,6))

xsamples = Finv(z,1)
    
hist(xsamples,50,normed = True,label = 'Samples distribution')
plot(x,f(x,lam),label='True distribution',lw = 3, color = 'r')

legend(loc = 0)
show()
```

###Exercise 2. Exponential Function - Numerical

```python
x1 =arange(0.,10.1,0.1)
lam = 1
f = lambda x,lam: lam*e**(-lam*x)
    
farr = f(x1, lam)

A = cumsum(farr)
B = sum(farr)
normA = A/B
#print normA

N = 100
z = numpy.random.random_sample(N);

arr = []
for q in z:
    arr = append(arr,searchsorted(normA,q))
print arr
```
